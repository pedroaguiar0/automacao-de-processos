# Automação de Sistemas e Processos com Python

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
Essa automação irá envia um email para diretoria diariamente, com o faturamento e a quantidade de produtos vendidos no dia anterior

Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing

Para resolver isso, usei o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado.

```python
import pyautogui
import pyperclip
import time

# Passo 01: Acessar sistema da empresa
pyautogui.PAUSE = 1

pyautogui.hotkey ("ctrl","t")
pyperclip.copy ("https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing")
pyautogui.hotkey ("ctrl","v")
pyautogui.press ("enter")
time.sleep (5)

# Passo 02: Localizar base de dados dentro do sistema
pyautogui.click (x=271, y=230, clicks=2)
time.sleep (3)

# Passo 03: Baixar base de dados
pyautogui.click (x=290, y=282)
pyautogui.click (x=1755, y=143)
pyautogui.click (x=1550, y=488)
time.sleep (5)
```

### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos

```python
# Passo 04: Importar base de dados para Python
import pandas as pd

tabela = pd.read_excel (r"C:\Users\Pedro\Downloads\Vendas - Dez.xlsx")
display (tabela)
```

```python
# Passo 05: Calcular base de dados
# faturamento
faturamento = tabela ["Valor Final"].sum()

# quantidade de vendas
quantidade = tabela ["Quantidade"].sum()

print(faturamento)
print(quantidade)
```

### Vamos agora enviar um e-mail pelo gmail

```python
# Passo 06: Abrir email
pyautogui.hotkey ("ctrl","t")
pyperclip.copy ("https://mail.google.com/mail/u/0/?tab=rm&ogbl#inbox")
pyautogui.hotkey ("ctrl","v")
pyautogui.press ("enter")
time.sleep (5)
# Passo 07: Abrir opção de envio:
pyautogui.click (x=71, y=154)
#Passo 08: Preencher email
pyperclip.copy ("aguiarcnc@gmail.com")
pyautogui.hotkey ("ctrl","v")
pyautogui.press ("tab")
pyperclip.copy ("Relatório de Vendas")
pyautogui.hotkey ("ctrl","v")
pyautogui.press ("tab")

texto = f"""Bom dia,

O faturamento de ontem foi: R${faturamento:,.2f}.
E a quantidade de vendas foi: {quantidade:,}.

Abraços,
Pedro Aguiar."""
pyperclip.copy (texto)
pyautogui.hotkey ("ctrl","v")
pyautogui.hotkey ("ctrl","Enter")
```
